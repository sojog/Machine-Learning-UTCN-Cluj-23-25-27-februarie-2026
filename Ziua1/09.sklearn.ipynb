```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apartamente = np.array([
   (35, 50_000), (45, 65_000),(50, 80_000), (55, 175_000), (55, 185_000), (60, 75_000), (65, 115_000), (75, 185_000), (80, 200_000), (85, 215_000)
])
suprafete = apartamente[:, 0]
preturi = apartamente[:, 1]
```

```python
plt.scatter(suprafete, preturi);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete
```

```python
suprafete_2d = suprafete.reshape(-1, 1) 
suprafete_2d
```

```python
model.fit(suprafete_2d, preturi)
```

```python
model.coef_, model.intercept_
```

```python
(np.float64(3180.4180418041806), np.float64(-57915.29152915292))
```

```python
model.predict([[70]])
```

```python
plt.scatter(suprafete, preturi);
plt.plot(suprafete, model.coef_ * suprafete + model.intercept_, c="black");
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
knn_model.fit(suprafete_2d, preturi)
```

```python
knn_model.predict([[70]])
```

```python
plt.scatter(suprafete, preturi);
plt.plot(suprafete, model.coef_ * suprafete + model.intercept_, c="black");

x = 70
plt.scatter(x, model.predict([[x]]), c="green")
plt.scatter(x, knn_model.predict([[x]]), c="red")
```
