```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("boston_data.csv", index_col=0)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
standard_scaler
```

```python
y = df [["medv"]]
type(y), type(df)
```

```python
x = df.drop("medv", axis=1)
x
```

```python
x_transformed_array = standard_scaler.fit_transform(x)
x_transformed_array.shape, type(x_transformed_array)
```

```python
df_scaled = pd.DataFrame(x_transformed_array)
df_scaled
```

```python
x_scaled = pd.DataFrame(x_transformed_array, columns=x.columns)
x_scaled
```

```python
x_scaled.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=101)
x_train.shape, x_test.shape, y_train.shape, y_test.shape 
```

```python
linear_model = LinearRegression()
linear_model
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
linear_model.fit(x_train, y_train)
```

```python
knn_model.fit(x_train, y_train)
```

```python
linear_model.score(x_test, y_test)
```

```python
knn_model.score(x_test, y_test)
```

```python
knn_scores = {}

for i in range(2, 11):
    inner_knn_model = KNeighborsRegressor(n_neighbors=i)
    inner_knn_model.fit(x_train, y_train)
    score = inner_knn_model.score(x_test, y_test)
    knn_scores[i] = score
```

```python
knn_scores
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(
    estimator = KNeighborsRegressor(),
    param_grid = {
        'n_neighbors': range(2, 11)
    }
)
grid_model 
```

```python
grid_model.fit(x_train, y_train)
```

```python
grid_model.score(x_test, y_test)
```

```python
grid_model.best_estimator_
```
