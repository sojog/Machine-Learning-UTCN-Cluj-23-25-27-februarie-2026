```python
import pickle
from PIL import Image
import numpy as np
```

```python
with open("model.pkl", "rb") as file_reader:
    model = pickle.load(file_reader)
```

```python
type(model)
```

```python
def citeste_imagine(path):
    return np.array(Image.open(path).convert("L")).flatten() / 16
```

```python
image = citeste_imagine("cifra_7.png")

model.predict([image])
```
