```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("categories_datasets/2_categories.csv")
df
```

```python
uncat_df = df[["coord_x","coord_y"]]
uncat_df
```

```python
plt.scatter(df["coord_x"], df["coord_y"])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters=2)
kmeans_model
```

```python
kmeans_model.fit(uncat_df)
```

```python
set(kmeans_model.predict(uncat_df))
```

```python
kmeans_model.predict(uncat_df)
```

```python
# uncat_df["predicted"] = kmeans_model.predict(uncat_df)
# uncat_df
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=kmeans_model.predict(uncat_df));
```
